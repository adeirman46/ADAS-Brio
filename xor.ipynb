```python
import numpy as np
import matplotlib.pyplot as plt

# XOR input and output
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
outputs = np.array([[0], [1], [1], [0]])

# Initialize weights and biases
w1 = np.random.randn(2, 4)  # Weights for input to hidden layer
b1 = np.random.randn(4)     # Biases for hidden layer
w2 = np.random.randn(4, 1)  # Weights for hidden layer to output
b2 = np.random.randn(1)     # Biases for output layer

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

def mse_loss(y_true, y_pred):
    return ((y_true - y_pred) ** 2).mean()

learning_rate = 0.01

# Training the network
for epoch in range(10000):
    # Forward pass
    z1 = np.dot(inputs, w1) + b1  # Input to hidden layer
    a1 = sigmoid(z1)              # Activation of hidden layer
    z2 = np.dot(a1, w2) + b2      # Hidden to output layer
    a2 = sigmoid(z2)              # Final output

    # Calculate loss
    loss = mse_loss(outputs, a2)

    # Backpropagation
    dloss_da2 = 2 * (a2 - outputs)  # Derivative of loss with respect to a2
    da2_dz2 = sigmoid_derivative(a2)  # Derivative of sigmoid with respect to z2
    dloss_dz2 = dloss_da2 * da2_dz2  # Gradient of loss w.r.t z2

    # Gradients for w2 and b2
    dw2 = np.dot(a1.T, dloss_dz2)
    db2 = np.sum(dloss_dz2, axis=0)

    # Backpropagation to hidden layer
    dz2_da1 = w2  # Derivative of z2 w.r.t a1
    dloss_da1 = dloss_dz2.dot(dz2_da1.T)  # Gradient w.r.t a1
    da1_dz1 = sigmoid_derivative(a1)  # Derivative of sigmoid w.r.t z1
    dloss_dz1 = dloss_da1 * da1_dz1  # Gradient of loss w.r.t z1

    # Gradients for w1 and b1
    dw1 = np.dot(inputs.T, dloss_dz1)
    db1 = np.sum(dloss_dz1, axis=0)

    # Update weights and biases
    w1 -= learning_rate * dw1
    b1 -= learning_rate * db1
    w2 -= learning_rate * dw2
    b2 -= learning_rate * db2

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Test the network after training
test_output = sigmoid(np.dot(sigmoid(np.dot(inputs, w1) + b1), w2) + b2)

# Plotting the output
plt.plot(test_output, label="Predicted Output")
plt.scatter(range(len(outputs)), outputs, color='red', label="True Output")
plt.legend()
plt.title("Predicted vs True Output for XOR")
plt.xlabel("Samples")
plt.ylabel("Output")
plt.show()
```

```python
import numpy as np

class Perceptron:
    # Inisialisasi nilai learning rate dan jumlah iterasi maksimum
    def __init__(self, learning_rate=0.01, max_iter=1000):
        self.learning_rate = learning_rate
        self.max_iter = max_iter
        self.weights = None
        self.bias = None
    
    def fit(self, X, y):
        # Mengambil jumlah sampel dan fitur dari data
        n_samples, n_features = X.shape
        
        # Inisialisasi bobot dan bias dengan nl
        self.weights = np.zeros(n_features)
        self.bias = 0
        
        # Looping sampai batas iterasi maksimum
        for epoch in range(self.max_iter):
            errors = 0  # Variabel error untuk menghitungg jumlah kesalahan dalam satu epoch/iterasi
            print(f"Epoch {epoch + 1}:")  # Menampilkan iterasi
            for idx, x_i in enumerate(X):
                # Menghitung output model berdasarkan weight (bobot) dan bias
                # Mengambil label target y yang sesuai dengan indeks saat ini (idx)
                update = y[idx] - self.step_function(np.dot(x_i, self.weights) + self.bias)
                
                # Jika prediksi salah, akan melakukan update pada weight dan bias
                if update != 0:  # Misclassified
                    self.weights -= self.learning_rate * update * x_i
                    self.bias -= self.learning_rate * update
                    errors += 1
                    print(f"  Update: weights={self.weights}, bias={self.bias}")
            
            # Jika tidak ada kesalahann mengklasifikasi, model akan berhenti
            print(f"  Total errors: {errors}\n")
            if errors == 0:
                break
    
    def step_function(self, x):
        return 1 if x >= 0 else 0  # Menggunakan Hard limit function

    def predict(self, X):
        return np.array([self.step_function(np.dot(x_i, self.weights) + self.bias) for x_i in X])

# Contoh penggunaan dengan hard limit
def main():
    X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])  # Input untuk AND gate
    y = np.array([0, 0, 0, 1])  # Output AND gate (Hard limit: 0 atau 1)

    # Inisialisasi model perceptron di atas
    model = Perceptron()
    model.fit(X, y)
    
    # Mencetak hasil weight dan bias setalah training
    print("Bobot akhir:", model.weights)
    print("Bias akhir:", model.bias)
    
    # Melakukan prediksi dengan model yang telah dilatih
    predictions = model.predict(X)
    print("Prediksi akhir:", predictions)
    
if __name__ == "__main__":
    main()
```
