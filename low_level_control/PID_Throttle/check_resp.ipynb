```python
import matplotlib.pyplot as plt

# Plot actual vs desired speed
plt.figure(figsize=(10, 6))
plt.plot(df["Time"], df["Desired_Speed"], label='Desired Speed', color='blue')
plt.plot(df["Time"], df["Actual_Speed"], label='Actual Speed', color='red')

# Set axis labels and title
plt.xlabel('Time (s)')
plt.ylabel('Speed (kph)')
plt.title('Speed Control Performance')

# Set y-axis to start from 0
plt.ylim(bottom=0)

# Add grid and legend
plt.grid(True)
plt.legend()

plt.show()
```
