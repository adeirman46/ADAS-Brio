## 10th order FIR

```python
# Load the data
dff = pd.read_csv('brake_data_with_time_fir_10th_order.csv')
# Subtract the first value of the 'Time (ms)' column from all the values in the column
dff['Time (ms)'] = dff['Time (ms)'] - dff['Time (ms)'].iloc[0]

# Save the updated dataframe to a new CSV file
dff.to_csv('brake_data_with_time_fir_10th_order_cleaned.csv', index=False)
```

```python
dfnf = pd.read_csv('brake_data_with_time_reset_no_filter_1.csv')
dff = pd.read_csv('brake_data_with_time_fir_10th_order_cleaned.csv')
dfi = pd.read_csv('brake_data_with_time_reset_filter_1.csv')
plt.plot(dfnf['Time (ms)'], dfnf['Distance (mm)'])
plt.plot(dff['Time (ms)'], dff['Distance (mm)'])
plt.plot(dfi['Time (ms)'], dfi['Distance (mm)'])
```

```python
dfnf = pd.read_csv('brake_data_with_time_reset_no_filter_1.csv')
dff = pd.read_csv('brake_data_with_time_fir_10th_order_cleaned.csv')
dfi = pd.read_csv('brake_data_with_time_reset_filter_1.csv')
dfi1 = pd.read_csv('brake_data_with_time_iir_1st_cleaned.csv')
plt.plot(dfnf['Time (ms)'], dfnf['Distance (mm)'])
plt.plot(dff['Time (ms)'], dff['Distance (mm)'])
plt.plot(dfi['Time (ms)'], dfi['Distance (mm)'])
plt.plot(dfi1['Time (ms)'], dfi1['Distance (mm)'])
# legend
plt.legend(['No Filter', 'FIR 10th Order', 'IIR 1st Order 1 Dec', 'IIR 1st Order 2 Dec'])
```
